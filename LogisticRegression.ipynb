```python
#Calculate age for missing values
def impute_age(cols):
    Age = cols[0]
    Pclass =cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train["Age"]=train[["Age","Pclass"]].apply(impute_age,axis=1)
```

```python
#Now we have no missing information for age column
sns.heatmap(train.isnull())
```

```python
#Drop cabin type since many missing information
```

```python
train.drop("Cabin", axis=1,inplace=True)
```

```python
#Drop any rows with missing information
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull())
```

```python
#Get categorical variables like Male/Female into dummy variables like 1 or 0 
#so that the machine learning algoithm can understand it
pd.get_dummies(train["Sex"])
```

```python
sex = pd.get_dummies(train["Sex"], drop_first=True)
```

```python
embark = pd.get_dummies(train["Embarked"], drop_first=True)
```

```python
train=pd.concat([train,sex,embark],axis=1)
```

```python
train.head(2)
```

```python
train.drop(["Sex","Embarked","Name","Ticket"], axis=1, inplace="True")
```

```python
train.drop("PassengerId",axis=1,inplace="True")
```

```python
train.head()
```

```python
X = train.drop("Survived", axis=1)
y = train["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
confusion_matrix(y_test,predictions)
```
